```python
# prompt: import csv file to use

import pandas as pd

# Change the path to point to the CSV file on your Google Drive
doc_df = pd.read_csv('/content/drive/My Drive/doc_table.csv')

# Print the DataFrame
doc_df
```

```python
import numpy as np
import pandas as pd

# Set random seed for reproducibility
np.random.seed(42)

# Generate random ratings for doctors according to different patients
num_doctors = 500
num_patients = 1000

# Generate doctor names
doctor_names = doc_df['Doctor Name']
patient_ids = [f"Patient_{i}" for i in range(1, num_patients + 1)]


# Generate random ratings matrix
ratings_matrix = np.random.choice([0, 1, 2, 3, 4, 5], size=(num_patients, num_doctors))

# Create DataFrame
ratings_df = pd.DataFrame(ratings_matrix, columns=doctor_names, index=patient_ids)

ratings_df
```
